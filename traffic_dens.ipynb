# TRAFFIC DENSITY ESTIMATION

## NECESSARY IMPORTS

```python
import cv2
import numpy as np
import matplotlib.pyplot as plt
```

## PATHS FOR YOLO MODEL

```python
# Define paths to YOLO model files
weight_path = "C:\\Users\\Sanjay\\Downloads\\yolov3.weights"
names_path = 'C:\\Users\\Sanjay\\Downloads\\coco.names'
cfg_path="C:\\Users\\Sanjay\\Downloads\\yolov3.cfg"
```

## TRAFFIC DENSITY CALCULATION

```python
def calculate_density(image):
    
    # Define the vehicle classes
    vehicle_classes = ['car', 'bus', 'truck', 'motorbike', 'bicycle']

    # Load YOLO model
    net = cv2.dnn.readNet(weight_path, cfg_path)
    layer_names = net.getLayerNames()
    output_layers = [layer_names[i - 1] for i in net.getUnconnectedOutLayers()]

    # Load COCO class labels
    with open(names_path, 'r') as f:
        classes = [line.strip() for line in f.readlines()]


    # Check if image is loaded
    if image is None:
        print(f"Error: Unable to load image at {image_path}")
        return

    height, width, channels = image.shape

    # Prepare the image for YOLO
    blob = cv2.dnn.blobFromImage(image, 0.00392, (416, 416), (0, 0, 0), True, crop=False)
    net.setInput(blob)

    # Run the YOLO model
    outs = net.forward(output_layers)

    # Process the detections
    class_ids = []
    confidences = []
    boxes = []

    for out in outs:
        for detection in out:
            scores = detection[5:]
            class_id = np.argmax(scores)
            confidence = scores[class_id]
            if confidence > 0.5 and classes[class_id] in vehicle_classes:
                center_x = int(detection[0] * width)
                center_y = int(detection[1] * height)
                w = int(detection[2] * width)
                h = int(detection[3] * height)
                x = int(center_x - w / 2)
                y = int(center_y - h / 2)
                boxes.append([x, y, w, h])
                confidences.append(float(confidence))
                class_ids.append(class_id)

    # Apply non-maxima suppression to remove redundant overlapping boxes
    indexes = cv2.dnn.NMSBoxes(boxes, confidences, 0.5, 0.4)

    # Initialize variables to calculate traffic density
    total_vehicle_area = 0
    num_vehicles = 0

    # Draw bounding boxes around detected vehicles
    for i in range(len(boxes)):
        if i in indexes:
            x, y, w, h = boxes[i]
            label = str(classes[class_ids[i]])
            color = (0, 255, 0)
            cv2.rectangle(image, (x, y), (x + w, y + h), color, 2)
            cv2.putText(image, label, (x, y - 10), cv2.FONT_HERSHEY_SIMPLEX, 0.5, color, 2)

            # Calculate the area of the bounding box and update the total vehicle area
            vehicle_area = w * h
            total_vehicle_area += vehicle_area
            num_vehicles += 1

    # Calculate traffic density
    total_image_area = width * height
    traffic_density = total_vehicle_area / total_image_area
    
    # Display the result
    plt.imshow(cv2.cvtColor(image, cv2.COLOR_BGR2RGB))
    plt.title("Traffic Density")
    plt.axis('off')
    plt.show()
    
    # Display the traffic density and number of vehicles
    print(f'Traffic Density: {traffic_density:.2%}')
    print(f'Number of Vehicles: {num_vehicles}')

    cv2.waitKey(0)
    cv2.destroyAllWindows()
```

## EXTRACT REGION OF INTEREST

```python
def extract_roi(frame, x_min, y_min, x_max, y_max):
    # Extract the ROI from the frame
    roi_image = frame[y_min:y_max, x_min:x_max]

    # Call the function to calculate density with the extracted ROI image
    if roi_image is not None:
        calculate_density(roi_image)
```

```python
def process_video(video_path):
    # Open the video file
    cap = cv2.VideoCapture(video_path)

    if not cap.isOpened():
        print("Error: Could not open video.")
        return

    # Get the frames per second (fps) of the video
    fps = cap.get(cv2.CAP_PROP_FPS)
    frame_interval = int(fps * 5)  # Interval to capture frames every 5 seconds

    frame_count = 0
    while True:
        ret, frame = cap.read()
        if not ret:
            break
        
        # Check if the current frame is at the 5-second interval
        if frame_count % frame_interval == 0:
            x_min, y_min = 300, 585
            x_max, y_max = 876, 1080
            extract_roi(frame, x_min, y_min, x_max, y_max)
        frame_count += 1

    cap.release()

# Example usage
video_path = "C:\\Users\\Sanjay\\Downloads\\3727445-hd_1920_1080_30fps.mp4"
process_video(video_path)
```
